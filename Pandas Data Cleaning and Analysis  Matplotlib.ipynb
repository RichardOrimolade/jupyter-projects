```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [0, 1 , 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.figure(figsize=(5,3), dpi=200)

# plt.plot(x, y, label = '2x', color = 'yellow', linewidth = 2, marker='.', markeredgecolor='red', markersize = 10, linestyle='--')

plt.plot(x,y, 'r.--', label='2x')

# we can use color picker code
# sift tab t bring out what is needed in your function () ie arange([start,] stop[, step,]

x2 = np.arange(0, 4.5, 0.5)

y2 = x2**2

plt.plot(x2[:6],y2[:6], 'b.--', label='2**x')
plt.plot(x2[5:],y2[5:], 'b.--',)

plt.title('First graph: Graph of x and y', fontdict={'fontname':'Comic Sans MS', 'fontsize':10})
plt.xlabel('Cost of Goods')
plt.ylabel('Price in USD')

plt.xticks([0, 1 , 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10, 12])

# insert legend

plt.legend()

plt.savefig('firstgraph2.png', dpi=300)

plt.show()
```

```python
# instead of writing all the followings what about writing single code for them all?

df = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(9,5))

plt.title('Gas Prce Over Time (US DOLLARS)', 
          fontdict={'fontweight': 'bold', 'fontsize': 14})

# plt.plot(df['Year'], df['Canada'], 'r.-', label='Canada')
# plt.plot(df['Year'], df['Germany'], 'b.-', label='Germany')
# plt.plot(df['Year'], df['Mexico'], 'y.-', label='Mexico')
# plt.plot(df['Year'], df['South Korea'], 'g.-', label= 'South Korea')

analysed_coutries = ['Canada', 'Germany', 'Mexico', 'South Korea', 'USA']
for country in df:
    if country in analysed_coutries:
        plt.plot(df['Year'], df[country], marker='.', label=country)

plt.xlabel("Year")
plt.ylabel("Price in USD")

plt.xticks(df['Year'][::5])
# the slicng above ::5 means the whole dataframe from start to ending by skiping it five years interval

plt.legend()

plt.savefig('gas_priceline1.png', dpi=300)

plt.show()
```
